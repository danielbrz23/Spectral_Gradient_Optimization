```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('TESTE.csv', index_col=0, header=[0, 1])
df =df.set_index(df.columns[0])
df.index.name = 'Problems'
```

```python
conj = [] 
for col in df.columns:
    met, _ = col
    conj.append(met)

conj = set(conj)

conj
```

```python
tempos =[ ] 
for c in conj:
    DF = df[c]
    tempos.append({c: round(DF['tempo (s)'].sum(), 3)})
tempos
```

```python
tempos_conv =[ ] 
for c in conj:
    DF = df[c]
    DF = DF[DF['evalf(x)']>0]
    tempos_conv.append({c: round(DF['tempo (s)'].sum(), 3)})
tempos_conv
```

```python
total_len = len(df)
conv= {}
for c in conj:
    DF = df[c]
    DF = DF[DF['evalf(x)']>=0]
    conv[c] = len(DF)
conv
```

```python
eval= []
for c in conj:
    DF = df[c]
    DF = DF[DF['evalf(x)']>0]
    eval.append( {'evalsf': DF['evalf(x)'].sum(), 'evalsgf': DF[' eval∇f(x)'].sum()})
d = pd.DataFrame(eval,index=list(conj))
d.sort_values(by =['evalsf'])
```
